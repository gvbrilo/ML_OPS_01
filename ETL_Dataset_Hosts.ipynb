```python
#STEP 6 : We normalize the format of the dates to: 'YYYY-MM-DD'
```

```python
#Step 6.a : Normalize the format of the 'date_added' column
#           for Dataset : Amazon
df_amazon['date_added'] = pd.to_datetime(df_amazon['date_added'])
df_amazon['date_added'] # Check the column 
```

```python
#Step 6.b : We normalize the format of the 'date_added' column
#           for Dataset : Disney
df_disney['date_added'] = pd.to_datetime(df_disney['date_added'])
df_disney['date_added'] # Check the column
```

```python
#Step 6.c : We normalize the format of the 'date_added' column
#           for Dataset : Hulu
df_hulu['date_added'] = pd.to_datetime(df_hulu['date_added'])
df_hulu['date_added'] # Check the column
```

```python
#Step 6.d : We normalize the format of the 'date_added' column
#           for Dataset : Netflix
df_netflix['date_added'] = pd.to_datetime(df_netflix['date_added'])
df_netflix['date_added'] # Check the column
```

```python
#STEP 7: We modify the 'txt' fields so that they are all in lower case
```

```python
#Step 7.a : We modify the 'txt' fields to lower case for the
#           Dataset : Amazon
df_amazon['type'] = df_amazon['type'].str.lower()
df_amazon['title'] = df_amazon['title'].str.lower()
df_amazon['director'] = df_amazon['director'].str.lower()
df_amazon['cast'] = df_amazon['cast'].str.lower()
df_amazon['country'] = df_amazon['country'].str.lower()
df_amazon['rating'] = df_amazon['rating'].str.lower()
df_amazon['duration'] = df_amazon['duration'].str.lower()
df_amazon['listed_in'] = df_amazon['listed_in'].str.lower()
df_amazon['description'] = df_amazon['description'].str.lower()
```

```python
#Step 7.b : We modify the 'txt' fields to lower case for the
#           Dataset : Disney
df_disney['type'] = df_disney['type'].str.lower()
df_disney['title'] = df_disney['title'].str.lower()
df_disney['director'] = df_disney['director'].str.lower()
df_disney['cast'] = df_disney['cast'].str.lower()
df_disney['country'] = df_disney['country'].str.lower()
df_disney['rating'] = df_disney['rating'].str.lower()
df_disney['duration'] = df_disney['duration'].str.lower()
df_disney['listed_in'] = df_disney['listed_in'].str.lower()
df_disney['description'] = df_disney['description'].str.lower()
```

```python
#Step 7.c : We modify the 'txt' fields to lower case for the
#           Dataset : Hulu
df_hulu['type'] = df_hulu['type'].str.lower()
df_hulu['title'] = df_hulu['title'].str.lower()
df_hulu['director'] = df_hulu['director'].str.lower()
df_hulu['country'] = df_hulu['country'].str.lower()
df_hulu['rating'] = df_hulu['rating'].str.lower()
df_hulu['duration'] = df_hulu['duration'].str.lower()
df_hulu['listed_in'] = df_hulu['listed_in'].str.lower()
df_hulu['description'] = df_hulu['description'].str.lower()
```

```python
#Step 7.d : We modify the 'txt' fields to lowercase for the
#           Dataset : Netflix
df_netflix['type'] = df_netflix['type'].str.lower()
df_netflix['title'] = df_netflix['title'].str.lower()
df_netflix['director'] = df_netflix['director'].str.lower()
df_netflix['cast'] = df_netflix['cast'].str.lower()
df_netflix['country'] = df_netflix['country'].str.lower()
df_netflix['rating'] = df_netflix['rating'].str.lower()
df_netflix['duration'] = df_netflix['duration'].str.lower()
df_netflix['listed_in'] = df_netflix['listed_in'].str.lower()
df_netflix['description'] = df_netflix['description'].str.lower()
```

```python
#STEP 8 : We divide the duration field into 2 ('duration_int' and 'duration_type'
```

```python
#Step 8.a : We divide the duration field in 2 for the Dataset : Amazon
df_amazon['duration'] = df_amazon['duration'].str.replace(r'', '')
df_amazon[['duration_int','duration_type']] = df_amazon['duration'].str.split(expand=True)
df_amazon #We check Amazon Dataset
```

```python
#Step 8.b : We divide the duration field in 2 for the Dataset : Disney 
df_disney['duration'] = df_disney['duration'].str.replace(r'', '')
df_disney[['duration_int','duration_type']] = df_disney['duration'].str.split(expand=True)
df_disney #We check Disney Dataset
```

```python
#Step 8.c : We divide the duration field in 2 for the Dataset : Hulu
df_hulu['duration'] = df_hulu['duration'].str.replace(r'', '')
df_hulu[['duration_int','duration_type']] = df_hulu['duration'].str.split(expand=True)
df_hulu #We check Hulu Dataset
```

```python
#Step 8.d : We divide the duration field in 2 for the Dataset: Neftlix
df_netflix['duration'] = df_netflix['duration'].str.replace(r'', '')
df_netflix[['duration_int','duration_type']] = df_netflix['duration'].str.split(expand=True)
df_netflix #We check Netflix Dataset
```

```python
#Step 9 : We create a Variable that contains the 4 unified datasets
df_completo = pd.concat([df_amazon , df_disney , df_hulu , df_netflix])
```

```python
#Step 10 : We transform the data type of the column "duration_int"
#          to integer:
df_completo['duration_int'] = df_completo['duration_int'].astype("Int64")

#Verify that the transformation has been done:
print(df_completo['duration_int'].dtype)
```

```python
#Step 11 : We review the data types in each column:
print(df_completo.dtypes)
```

```python
#Step 12 : We remove the columns that are redundant
df_completo = df_completo.drop(columns= ["show_id", "duration"])
```

```python
#Step 13 : We review the final dataset
df_completo
```

```python
#Step 14: We export the file to a new CSV:
df_completo.to_csv("Datasets_Hosts/df_total_hosts.csv", index=False)
```
