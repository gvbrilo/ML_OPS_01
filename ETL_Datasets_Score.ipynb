```python
#We import the libraries that we will use
import pandas as pd
```

```python
#Step 1 : Define our csv's
df_1 = pd.read_csv('Datasets_Score/1.csv')

df_2 = pd.read_csv('Datasets_Score/2.csv')

df_3 = pd.read_csv('Datasets_Score/3.csv')

df_4 = pd.read_csv('Datasets_Score/4.csv')

df_5 = pd.read_csv('Datasets_Score/5.csv')

df_6 = pd.read_csv('Datasets_Score/6.csv')

df_7 = pd.read_csv('Datasets_Score/7.csv')

df_8 = pd.read_csv('Datasets_Score/8.csv')

```

```python
#Step 2 : Concatenating our dataframes to from df_total_score 
df_total_score = pd.concat([df_1  , df_2 , df_3 , df_4 , df_5 , df_6 , df_7 , df_8  ])
```

```python
df_total_score
```

```python
#Step 3 : Calculate the number of duplicated rows in the DataFrame
df_total_score.duplicated().sum()
```

```python
#Step 4 : Remove duplicate rows from the DataFrame 
df_total_score = df_total_score.drop_duplicates()

df_total_score.duplicated().sum()
```

```python
#Step 4 : Change the format of the 'timestamp' column 
df_total_score['timestamp'] = pd.to_datetime(df_total_score['timestamp'], unit='s').dt.strftime('%Y - %m - %d')

#Step 5 : Turn in to date format
df_total_score['timestamp'] = pd.to_datetime(df_total_score['timestamp'])
```

```python
df_total_score
```

```python
#Step 6 : We export the Dataframe to a new CSV (as well as this will serve us
#         later for the EDA part and ML training):

df_total_score.to_csv("Datasets_Score/df_total_score.csv", index=False)
```

```python
# Step 7 : calculate the mean of the 'rating' row and transform it so we can  join with 
#          df_dataset_hosts dataframe 

df_score = df_total_score.groupby('movieId')['rating','movieId'].mean().round(2).reset_index()

df_score = df_score.rename(columns={'rating': 'score_prom'})

df_score
```

```python
#Step 8 : Define our df_total_hosts csv
df_dataset_hosts = pd.read_csv('Datasets_Hosts/df_total_hosts.csv')

df_dataset_hosts
```

```python
#Step 9 : Join the df_score with the df_dataset_hosts using the 'movieId' column as the 
#         key column for both DataFrames.

df_total_data = df_score.join(df_dataset_hosts.set_index('id'), on='movieId')

df_total_data
```

```python
# Step 10 : Add a new column 'platform' to df_total_data , the value of this column is calculated 
#           by replacing specific substrings  in the 'movieId' column using regular expressions 

df_total_data['platform'] = df_total_data['movieId'].str.replace(r'as\d+', 'amazon' ).str.replace(r'ds\d+', 'disney').str.replace(r'hs\d+', 'hulu').str.replace(r'ns\d+', 'netflix')

df_total_data
```

```python
#Step 11 : We export the final analysis Dataframe for the queries
df_total_data.to_csv("df_total_data.csv", index=False)
```
